```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials
import pandas as pd
import numpy as np
import time

scope = ['https://spreadsheets.google.com/feeds','https://www.googleapis.com/auth/drive']
creds = ServiceAccountCredentials.from_json_keyfile_dict(create_keyfile_dict(), scope)
client = gspread.authorize(creds)
spreadsheets = ['test_feedback_responces']

#Open the Spreadsheet or several Spreadsheets
for spreadsheet in spreadsheets:
    sh = client.open(spreadsheet)

#Get all values in the first worksheet
worksheet = sh.get_worksheet(0)
data = worksheet.get_all_values()

#Save the data inside the pandas dataframe
df = pd.DataFrame(columns = (data[0]))
for i in range(1,len(data)):
    df.loc[len(df)] = data[i]      
```

```python
df = df.replace(r'^\s*$', np.nan, regex=True)
df
```

```python
#dataframe with likert scale questions only

dff = df.iloc[:, [0, 6, 7, 9, 10, 11, 17, 18, 19]].copy()

dff.set_index('Timestamp', inplace=True)

dff
```

```python
df_count = dff.apply(pd.Series.value_counts)
df_count = df_count.fillna(0).astype(int)
df_count
```

```python
list_of_counts = df_count.transpose().values.tolist()
```

```python
import matplotlib.pyplot as plt


category_names = ['Trifft nicht zu', 'Trifft teilweise nicht zu',
                  'Neutral', 'Trifft teilweise zu', 'Trifft zu']

#enumerate(category_names, start=1)
#Name_list = df["<question>"].tolist()





questions = list(df_count.columns)

results = dict(zip(questions, list_of_counts))


def survey(results, category_names):
    """
    Parameters
    ----------
    results : dict
        A mapping from question labels to a list of answers per category.
        It is assumed all lists contain the same number of entries and that
        it matches the length of *category_names*.
    category_names : list of str
        The category labels.
    """
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.colormaps['RdYlGn'](
        np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(9.2, 5))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        rects = ax.barh(labels, widths, left=starts, height=0.5,
                        label=colname, color=color)

        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        ax.bar_label(rects, label_type='center', color=text_color)
    ax.legend(ncol=len(category_names), bbox_to_anchor=(0, 1),
              loc='lower left', fontsize='small')

    return fig, ax


survey(results, category_names)

plt.show()
```

```python
'''Two test cases: test_output_type and test_output_content. 
    The test_output_type test case checks if the returned objects are
    of the correct type (matplotlib.Figure and matplotlib.Axes). 
    The test_output_content test case checks if the expected number of containers exist in the Axes object.'''

import unittest

class TestSurveyFunction(unittest.TestCase):
    def test_output_type(self):
        results = dict(zip(questions, list_of_counts))
        category_names = ['Trifft nicht zu', 'Trifft teilweise nicht zu',
                  'Neutral', 'Trifft teilweise zu', 'Trifft zu']
        fig, ax = survey(results, category_names)
        self.assertIsInstance(fig, plt.Figure)
        self.assertIsInstance(ax, plt.Axes)

    def test_output_content(self):
        results = dict(zip(questions, list_of_counts))
        category_names = ['Trifft nicht zu', 'Trifft teilweise nicht zu',
                  'Neutral', 'Trifft teilweise zu', 'Trifft zu']
        fig, ax = survey(results, category_names)
        self.assertEqual(len(ax.containers), 5)

if __name__ == '__main__':
    unittest.main(argv=['ignored', '-v'], exit=False)
```

```python
from matplotlib import cm, gridspec
import math
from PIL import Image
from mpl_toolkits.axes_grid1 import make_axes_locatable
 
# set your color array and name of figure here:
dial_colors = np.linspace(0,1,1000) # using linspace here as an example
figname = 'myDial'
 
# specify which index you want your arrow to point to
arrow_index = round(time_mean._get_value('Ich hatte KEINEN Termin', 'Wartezeit (Minuten)'), 1)*10 
 
# create labels at desired locations
# note that the pie plot ploots from right to left
labels = [' ']*len(dial_colors)*2
labels[25] = '100'
labels[250] = '75'
labels[500] = '50'
labels[750] = '25'
labels[975] = '0'
 
# function plotting a colored dial
def dial(color_array, arrow_index, labels, ax):
    # Create bins to plot (equally sized)
    size_of_groups=np.ones(len(color_array)*2)
 
    # Create a pieplot, half white, half colored by your color array
    white_half = np.ones(len(color_array))*.5
    color_half = color_array
    color_pallet = np.concatenate([color_half, white_half])
 
    cs=cm.RdYlGn(color_pallet)
    pie_wedge_collection = ax.pie(size_of_groups, colors=cs, labels=labels)
 
    i=0
    for pie_wedge in pie_wedge_collection[0]:
        pie_wedge.set_edgecolor(cm.RdYlGn(color_pallet[i]))
        i=i+1
 
    # create a white circle to make the pie chart a dial
    my_circle=plt.Circle( (0,0), 0.3, color='white')
    ax.add_artist(my_circle)
 
    # create the arrow, pointing at specified index
    arrow_angle = (arrow_index/float(len(color_array)))*3.14159
    arrow_x = 0.2*math.cos(arrow_angle)
    arrow_y = 0.2*math.sin(arrow_angle)
    ax.arrow(0,0,-arrow_x,arrow_y, width=.02, head_width=.05, \
        head_length=.1, fc='k', ec='k')
 
# create figure and specify figure name
fig, ax = plt.subplots()
 
# make dial plot and save figure
dial(dial_colors, arrow_index, labels, ax)
ax.set_aspect('equal')
plt.savefig(figname + '.png', bbox_inches='tight') 
 
 
# open figure and crop bottom half
im = Image.open(figname + '.png')
width, height = im.size

# crop bottom half of figure
im1 = im.crop((0, 0, width, int(height/2.0)))#.save(figname + '.png')
im1.show()
```
